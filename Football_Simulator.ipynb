```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
EPL_data = pd.read_csv('E0.csv', delimiter = ',')
```

```python
EPL_data
```

```python
home_teams = list(set(list(EPL_data['HomeTeam'])))
away_teams = list(set(list(EPL_data['AwayTeam'])))
```

```python
#Checking how many goals each team scores on average at home
avg_home_goals = np.zeros(len(home_teams))
AVG_HG = dict(zip(home_teams, avg_home_goals))

for i in range(EPL_data.shape[0]):
    AVG_HG[EPL_data['HomeTeam'][i]] += EPL_data['FTHG'][i]

        
for key, value in AVG_HG.items():
    AVG_HG[key] = value/19
```

```python
AVG_HG
```

```python
#Checking how many goals each team scores on average away
avg_away_goals = np.zeros(len(home_teams))
AVG_AG = dict(zip(home_teams, avg_away_goals))

for i in range(EPL_data.shape[0]):
    AVG_AG[EPL_data['AwayTeam'][i]] += EPL_data['FTAG'][i]

        
for key, value in AVG_AG.items():
    AVG_AG[key] = value/19
```

```python
AVG_AG
```

```python
#Average Shots for Home team
avg_home_shots = np.zeros(len(home_teams))
AVG_HS = dict(zip(home_teams, avg_home_shots))

for i in range(EPL_data.shape[0]):
#     print(EPL_data['HomeTeam'][i], EPL_data['HS'][i])
    AVG_HS[EPL_data['HomeTeam'][i]] += EPL_data['HS'][i]

        
for key, value in AVG_HS.items():
    AVG_HS[key] = value/19
```

```python
AVG_HS
```

```python
# Average Away Team Shots
avg_away_shots = np.zeros(len(home_teams))
AVG_AS = dict(zip(home_teams, avg_away_shots))

for i in range(EPL_data.shape[0]):
#     print(EPL_data['HomeTeam'][i], EPL_data['HS'][i])
    AVG_AS[EPL_data['AwayTeam'][i]] += EPL_data['AS'][i]

        
for key, value in AVG_AS.items():
    AVG_AS[key] = value/19
```

```python
def get_average(team, column):
    """Gets average of a statistic passed to this function for the Home or Away Team"""
    avg_stat = np.zeros(len(home_teams))
    AVG_DICT = dict(zip(home_teams, avg_stat))
    for i in range(EPL_data.shape[0]):
        AVG_DICT[EPL_data[team][i]] += EPL_data[column][i]

        
    for key, value in AVG_DICT.items():
        AVG_DICT[key] = value/19
        
    return AVG_DICT
```

```python
EPL_data.columns
```

```python
home_stats = ['HTHG','FTHG','HS', 'HST', 'HC', 'HF', 'HY', 'HR']
away_stats = ['HTAG','FTAG','AS', 'AST', 'AC', 'AF', 'AY', 'AR']
teams = ['HomeTeam', 'AwayTeam']
```

```python
all_home_stats = {}
all_away_stats = {}
for team in teams:
    print("Printing " + team + " stats")
    if team == 'HomeTeam':
        for stat in home_stats:
            print("Average "+ stat)
            AVG_DICT = get_average(team, stat)
            all_home_stats[stat] = AVG_DICT
            print(AVG_DICT)
    else:
        for stat in away_stats:
            print("Average "+ stat)
            AVG_DICT = get_average(team, stat)
            all_away_stats[stat] = AVG_DICT
            print(AVG_DICT)
```
